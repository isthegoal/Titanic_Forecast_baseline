# 前置阶段：数据分析
对于大数据量，通过数据分析来更好的观察数据的分布变化，这样能有利找出关联性强的特征。

```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame #用padas做科学计算，numpy做数据分析
data=pd.read_csv("G:\Machine-learning\\tensorflow-train-project(jupyter)\kaggele-project6-tairanprediction\dataset\\train.csv")
data
```

```python
data.info()
```

```python
#属性分析阶段
import matplotlib.pyplot as plt
fig=plt.figure()
fig.set(alpha=0.2)#alpha=0.2   用于设定所创建图标的颜色
# 第一个柱型图  用于展示获救情况，  横向比是2   纵向比是3
plt.subplot2grid((2,3),(0,0))#在大图上创建小图，
data.Survived.value_counts().plot(kind='bar')#  创建一个柱状图，    这里会统计统一数值的数量，作为图表上的说明    纵向的表示人数
plt.title(u"the 1 is  survived")
plt.ylabel(u"the people num")

#第二个柱形图
plt.subplot2grid((2,3),(0,1))
data.Pclass.value_counts().plot(kind='bar')
plt.ylabel(u"the peopele number")
plt.title('the Pclass  fenbu')

#第三个散点图
plt.subplot2grid((2,3),(0,2))
plt.scatter(data.Survived,data.Age)# 创建散点图，并设定横纵坐标值填充，横坐标表示  获救情况的数值分布，纵轴是乘客年龄的分布
plt.ylabel(u'age')
plt.grid(b=True,which='major',axis='y')  #  这里是对展示网格的设置，只显示纵向网格，而不会交叉的
plt.title('see it from age,1 is survived')

#第四个是密度图         用于不同等级乘客的年龄分布情况，这里关联的属性是Pclass和Age
plt.subplot2grid((2,3),(1,0),colspan=2)#
data.Age[data.Pclass==1].plot(kind='kde')# 对客舱等级为1的设置为一种密度线
data.Age[data.Pclass==2].plot(kind='kde')# 对客舱等级为2的设置为一种密度线   这是一种关联性的计算 Age中不同的类别贴上对应的标签
data.Age[data.Pclass==3].plot(kind='kde')# 对客舱等级为3的设置为一种密度线
plt.xlabel(u'age')
plt.ylabel(u'midu')
plt.title('the different level people age fenbu')
plt.legend((u'tou deng',u'two deng',u'san deng'),loc='best')


#第五个图用于展示各登岸口上人数量分布
plt.subplot2grid((2,3),(1,2))
data.Embarked.value_counts().plot(kind='bar')#用直方图展示  不同登陆口的数量分布
plt.title('different bark people')
plt.ylabel(u'the people number')#u应该是说utf-8,可以没起效果
plt.show()
```

# 关联性分析
上面只是图表观察，接下来进行关联性的分析，挖掘更有用的属性关联

```python
#展示 乘客等级和获救之间的关联性
fig=plt.figure()#创建图表
fig.set(alpha=0.2)
#分别获取不同分类中   不同获救情况的数量
Survived_0=data.Pclass[data.Survived==0].value_counts()
Survived_1=data.Pclass[data.Survived==1].value_counts()
df=pd.DataFrame({'Survived':Survived_1,'unSurvived':Survived_0})#使用dataframe表示组成一个表格
df.plot(kind='bar',stacked=True)   #使用构造的dataframe来构造一个直方图数据分布，这样来填充展示
plt.title('different  passenger level survived situation')
plt.xlabel('passenger level')
plt.ylabel('the number')
plt.show()#一个表单位的效果展示
```

```python
#不同性别的获取情况分析   （可以对图看出性别很重要，男女不平等啊）
fig=plt.figure()
fig.set(alpha=0.9)#设定图标颜色值  透明度
#方法同上标记    这是一种计数的套路
survived_m=data.Survived[data.Sex=='male'].value_counts()
survived_f=data.Survived[data.Sex=='female'].value_counts()
df=pd.DataFrame({'male:':survived_m,'female':survived_f})
df.plot(kind='bar',stacked=True)
plt.title('different  sex survived situation')
plt.xlabel('gender')
plt.ylabel('survived')
plt.show()#第二图展示
```

```python
 #然后我们再来看看各种舱级别情况下各性别的获救情况    生成一个大图，得到不同获救和高低仓的组合情况。
fig=plt.figure(figsize=[15,15])  #调下尺寸
fig.set(alpha=0.65) # 设置图像透明度，无所谓
plt.title('the cang level with different gender  survived')

ax1=fig.add_subplot(141)  #subplot表示在大图中添加的子图，这四个图会合成一个大图
#组合形式的计数，得到女性非低级仓的获救计数   并指定横轴和纵轴意义
data.Survived[data.Sex == 'female'][data.Pclass != 3].value_counts().plot(kind='bar', label="female highclass", color='#FA2479')
ax1.set_xticklabels([u"survived", u"unsurvived"], rotation=0)
ax1.legend([u"female/high"], loc='best')#颜色标记

ax2=fig.add_subplot(142, sharey=ax1)
data.Survived[data.Sex == 'female'][data.Pclass == 3].value_counts().plot(kind='bar', label='female, low class', color='pink')
ax2.set_xticklabels([u"unsurvived", u"survived"], rotation=0)
plt.legend([u"female/low"], loc='best')

ax3=fig.add_subplot(143, sharey=ax1)
data.Survived[data.Sex == 'male'][data.Pclass != 3].value_counts().plot(kind='bar', label='male, high class',color='lightblue')
ax3.set_xticklabels([u"unsurvived", u"survived"], rotation=0)
plt.legend([u"male/high"], loc='best')

ax4=fig.add_subplot(144, sharey=ax1)
data.Survived[data.Sex == 'male'][data.Pclass == 3].value_counts().plot(kind='bar', label='male low class', color='steelblue')
ax4.set_xticklabels([u"unsurvived", u"survived"], rotation=0)
plt.legend([u"male/low"], loc='best')

plt.show()
```

```python
#不同登陆港口的获救情况
fig=plt.figure()#画板上创建一个新图
fig.set(alpha=0.2)
Survived_0=data.Embarked[data.Survived==0].value_counts()#不同登岸口  未获救的计数
Survived_1=data.Embarked[data.Survived==1].value_counts()#不同登岸口  未获救的计数
df=pd.DataFrame({'Survived':Survived_1,'unSurvived':Survived_0})
df.plot(kind='bar',stacked=True)
plt.title('different bar  Survived situation')
plt.xlabel('bar')
plt.ylabel('number')
plt.show()
```
